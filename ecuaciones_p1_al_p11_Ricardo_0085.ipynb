Bloque 2: Variables en ambos lados de la ecuaciónPara estos casos, el objetivo es "juntar" las letras de un lado y los números del otro.


* Problema 1: $x + 8 = 15$Planteamiento: Un número desconocido sumado a 8 da 15.Despeje: El 8 que suma pasa restando: $x = 15 - 8$.Solución: $x = 7$.

```python
# --- Problema: x + 8 = 15 ---
x = 15 - 8
print(f"Solución x + 8 = 15: x = {x}")
```

Problema 3 : $3x + 4x = 21$Planteamiento: Tres veces un número más cuatro veces ese mismo número es 21.Simplificación: Agrupamos términos semejantes: $(3+4)x = 21$, lo que nos da $7x = 21$.Despeje: El 7 que multiplica pasa dividiendo: $x = \frac{21}{7}$.Solución: $x = 3$.

```python
# --- Problema: 3x + 4x = 21 ---
# 7x = 21 -> x = 21 / 7
x2 = 21 / (3 + 4)
print(f"Solución 3x + 4x = 21: x = {int(x2)}")
```

Problema 5: $6a - 2a + a$ (Expresión)Nota: Al no tener un signo de igual (=), esto es una simplificación de expresión.Análisis: Tenemos 6 unidades de "a", restamos 2 y sumamos 1 (recordando que $a$ es lo mismo que $1a$).Operación: $6 - 2 + 1 = 5$.Resultado: $5a$.

```python
# --- Problema: 6a - 2a + a ---
# Simplificación: (6 - 2 + 1)a
coeficiente_a = 6 - 2 + 1
print(f"Simplificación 6a - 2a + a: {coeficiente_a}a")
```

Problema 7: $4x + 3 = 2x + 11$Agrupar variables: Pasamos $2x$ al lado izquierdo restando: $4x - 2x + 3 = 11 \rightarrow 2x + 3 = 11$.Agrupar números: Pasamos el 3 al lado derecho restando: $2x = 11 - 3 \rightarrow 2x = 8$.Despejar: $x = \frac{8}{2}$.Solución: $x = 4$.

```python
# --- Problema: 4x + 3 = 2x + 11 ---
# (4x - 2x) = 11 - 3 -> 2x = 8
x3 = (11 - 3) / (4 - 2)
print(f"Solución 4x + 3 = 2x + 11: x = {int(x3)}")
```

Problema 9: $9m + 4 = 5m + 20$Agrupar variables: $9m - 5m = 20 - 4$.Simplificar: $4m = 16$.Despejar: $m = \frac{16}{4}$.Solución: $m = 4$.

```python
# --- Problema: 9m + 4 = 5m + 20 ---
# (9m - 5m) = 20 - 4 -> 4m = 16
m = (20 - 4) / (9 - 5)
print(f"Solución 9m + 4 = 5m + 20: m = {int(m)}")
```
