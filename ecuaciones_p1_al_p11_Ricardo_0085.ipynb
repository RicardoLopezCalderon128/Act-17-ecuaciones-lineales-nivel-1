* Problema 1: $x + 8 = 15$Planteamiento: Un número desconocido sumado a 8 da 15.Despeje: El 8 que suma pasa restando: $x = 15 - 8$.Solución: $x = 7$.

```python
# Problema 1: x + 8 = 15
print("--- Resolviendo x + 8 = 15 ---")
b = float(input("Ingresa el resultado (15): "))
a = float(input("Ingresa el número que suma (8): "))
x = b - a
print(f"Resultado: x = {x}")
```

Problema 3 : $3x + 4x = 21$Planteamiento: Tres veces un número más cuatro veces ese mismo número es 21.Simplificación: Agrupamos términos semejantes: $(3+4)x = 21$, lo que nos da $7x = 21$.Despeje: El 7 que multiplica pasa dividiendo: $x = \frac{21}{7}$.Solución: $x = 3$.

```python
# Problema 2: 3x + 4x = 21
print("\n--- Resolviendo 3x + 4x = 21 ---")
c1 = float(input("Primer coeficiente (3): "))
c2 = float(input("Segundo coeficiente (4): "))
res = float(input("Resultado total (21): "))
x = res / (c1 + c2)
print(f"Resultado: x = {x}")
```

Problema 5: $6a - 2a + a$ (Expresión)Nota: Al no tener un signo de igual (=), esto es una simplificación de expresión.Análisis: Tenemos 6 unidades de "a", restamos 2 y sumamos 1 (recordando que $a$ es lo mismo que $1a$).Operación: $6 - 2 + 1 = 5$.Resultado: $5a$.

```python
# Problema 3: 6a - 2a + a
print("\n--- Simplificando 6a - 2a + a ---")
n1 = float(input("Coeficiente 1 (6): "))
n2 = float(input("Coeficiente 2 (2): "))
n3 = float(input("Coeficiente 3 (1): "))
resultado = n1 - n2 + n3
print(f"Resultado simplificado: {resultado}a")
```

Problema 7: $4x + 3 = 2x + 11$Agrupar variables: Pasamos $2x$ al lado izquierdo restando: $4x - 2x + 3 = 11 \rightarrow 2x + 3 = 11$.Agrupar números: Pasamos el 3 al lado derecho restando: $2x = 11 - 3 \rightarrow 2x = 8$.Despejar: $x = \frac{8}{2}$.Solución: $x = 4$.

```python
# Problema 4: 4x + 3 = 2x + 11
print("\n--- Resolviendo 4x + 3 = 2x + 11 ---")
a1 = float(input("Coeficiente x izquierdo (4): "))
n1 = float(input("Número izquierdo (3): "))
a2 = float(input("Coeficiente x derecho (2): "))
n2 = float(input("Número derecho (11): "))

x = (n2 - n1) / (a1 - a2)
print(f"Resultado: x = {x}")
```

Problema 9: $9m + 4 = 5m + 20$Agrupar variables: $9m - 5m = 20 - 4$.Simplificar: $4m = 16$.Despejar: $m = \frac{16}{4}$.Solución: $m = 4$.

```python
# Problema 5: 9m + 4 = 5m + 20
print("\n--- Resolviendo 9m + 4 = 5m + 20 ---")
m1 = float(input("Coeficiente m izquierdo (9): "))
c1 = float(input("Número izquierdo (4): "))
m2 = float(input("Coeficiente m derecho (5): "))
c2 = float(input("Número derecho (20): "))

m = (c2 - c1) / (m1 - m2)
print(f"Resultado: m = {m}")
```

Problema 11: $6a + 5 = 2a + 21$Agrupar variables: $6a - 2a = 21 - 5$.Simplificar: $4a = 16$.Despejar: $a = \frac{16}{4}$.Solución: $a = 4$.

```python
# Problema 6: 6a + 5 = 2a + 21
print("\n--- Resolviendo 6a + 5 = 2a + 21 ---")
coef1 = float(input("Coeficiente a izquierdo (6): "))
num1 = float(input("Número izquierdo (5): "))
coef2 = float(input("Coeficiente a derecho (2): "))
num2 = float(input("Número derecho (21): "))

a = (num2 - num1) / (coef1 - coef2)
print(f"Resultado: a = {a}")
```
